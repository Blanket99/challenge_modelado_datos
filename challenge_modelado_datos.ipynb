### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

url1 = "https://raw.githubusercontent.com/Blanket99/challenge_modelado_datos/refs/heads/main/base-de-datos/tienda_1.csv"
url2 = "https://raw.githubusercontent.com/Blanket99/challenge_modelado_datos/refs/heads/main/base-de-datos/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/Blanket99/challenge_modelado_datos/refs/heads/main/base-de-datos/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/Blanket99/challenge_modelado_datos/refs/heads/main/base-de-datos/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

print("Tienda 1")
display(tienda1.head())
print("\nTienda 2")
display(tienda2.head())
print("\nTienda 3")
display(tienda3.head())
print("\nTienda 4")
display(tienda4.head())
```

#1. Análisis de facturación



```python
# Agregar una columna nueva a cada DataFrame para identificar a qué tienda pertenece cada fila
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Combinar todos los DataFrames individuales en uno solo usando pd.concat()
# Se usa ignore_index=True para que los índices se reenumeren automáticamente
df_unificado = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Mostrar información general del DataFrame unificado (columnas, tipos, nulos, etc.)
print(df_unificado.info())

# Contar cuántas filas hay por tienda (útil para verificar distribución)
display(df_unificado['Tienda'].value_counts())
```

```python
# Agrupar por 'Tienda', sumar la columna 'Precio' y ordenar de mayor a menor
ingresos_por_tienda = df_unificado.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)
print("Ingresos Totales por Tienda:")
display(ingresos_por_tienda.reset_index(name='Ingresos'))
```

```python
# Gráfico de barras para ingresos por tienda
plt.figure(figsize=(8, 5))

# Crear el gráfico de barras con barras más delgadas
ax = sns.barplot(x=ingresos_por_tienda.index, y=ingresos_por_tienda.values, palette="pastel", hue=ingresos_por_tienda.index, width=0.5)

# Título y etiquetas
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos (en $)')
plt.xlabel('Tienda')

# Añadir etiquetas con el valor sobre cada barra
for i, v in enumerate(ingresos_por_tienda.values):
    ax.text(i, v + 0.1, f'{v:.1f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Contar cuántas veces se vendió cada categoría en cada tienda
categorias_vendidas = df_unificado.groupby(['Tienda', 'Categoría del Producto'])['Producto'].count()

# Para cada tienda, seleccionar la categoría con más ventas
categorias_mas_vendidas = categorias_vendidas.groupby(level=0).idxmax().apply(lambda x: x[1])
print("Categoría más vendida por Tienda:")
display(categorias_mas_vendidas.reset_index(name='Categoría'))
```

```python
# Obteniendo las categorías más vendidas por tienda
categorias_por_tienda = categorias_vendidas.groupby(['Tienda', 'Categoría del Producto']).sum().unstack().fillna(0)

# Graficar un gráfico circular por cada tienda
for tienda in categorias_por_tienda.index:
    # Obtener los datos de la tienda
    data = categorias_por_tienda.loc[tienda]

    # Crear el gráfico circular
    plt.figure(figsize=(5.5, 5.5))
    ax = data.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("Set3"), legend=False, pctdistance=0.75)

    # Poner en negrita la categoría con más ventas
    max_category = data.idxmax()
    max_value = data.max()

    # Acceder a los textos del gráfico (las etiquetas de categorías y los porcentajes)
    for i, text in enumerate(ax.texts):
        # Si el texto corresponde a la categoría más vendida, lo ponemos en negrita
        if max_category in text.get_text():
            text.set_fontweight('bold')

        # Revisamos si el texto es el porcentaje correspondiente a la categoría más vendida
        if f'{(max_value / data.sum()) * 100:.1f}%' in text.get_text():
            text.set_fontweight('bold')

        # Reducir el tamaño de las etiquetas y porcentajes
        text.set_fontsize(9)

    # Eliminar el título "Tienda X" que aparece como parte de las etiquetas
    ax.set_ylabel('')  # Eliminar la etiqueta del eje y

    # Título del gráfico
    plt.title(f'Categorías Más Vendidas - {tienda}', fontsize=12)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificaciones de clientes por tienda, redondeado a 2 decimales
calificaciones_promedio = df_unificado.groupby('Tienda')['Calificación'].mean().round(2)
envio_promedio = df_unificado.groupby('Tienda')['Costo de envío'].mean().round(2)
print("Calificación promedio por Tienda:")
display(calificaciones_promedio.reset_index(name='Calificación Promedio'))
```

```python
# Preparar datos combinados
datos_comparacion = pd.DataFrame({
    'Tienda': calificaciones_promedio.index,
    'Calificación Promedio': calificaciones_promedio.values,
    'Envío Promedio': envio_promedio.values
})

plt.figure(figsize=(8, 5))
sns.scatterplot(data=datos_comparacion,
                x='Envío Promedio',
                y='Calificación Promedio',
                hue='Tienda',
                s=100,
                palette='deep')
plt.title('Relación: Costo de Envío vs Calificación por Tienda')
plt.xlabel('Costo Promedio de Envío')
plt.ylabel('Calificación Promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Contar cuántas veces se vendió cada producto en cada tienda
productos_vendidos = df_unificado.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad Vendida')

# Obtener el producto más vendido por tienda
productos_mas_vendidos = productos_vendidos.loc[productos_vendidos.groupby('Tienda')['Cantidad Vendida'].idxmax()].reset_index(drop=True)
print("Producto más vendido por Tienda:")
display(productos_mas_vendidos)

# Obtener el producto menos vendido por tienda
productos_menos_vendidos = productos_vendidos.loc[productos_vendidos.groupby('Tienda')['Cantidad Vendida'].idxmin()].reset_index(drop=True)
print("\nProducto menos vendido por Tienda:")
display(productos_menos_vendidos)
```

```python
# Obtener los productos más y menos vendidos por tienda con sus cantidades
productos_vendidos = df_unificado.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad Vendida')

productos_mas_vendidos = productos_vendidos.loc[productos_vendidos.groupby('Tienda')['Cantidad Vendida'].idxmax()].reset_index(drop=True)
productos_mas_vendidos['Tipo'] = 'Más Vendido'

productos_menos_vendidos = productos_vendidos.loc[productos_vendidos.groupby('Tienda')['Cantidad Vendida'].idxmin()].reset_index(drop=True)
productos_menos_vendidos['Tipo'] = 'Menos Vendido'

# Combinar ambos en un solo DataFrame
productos_extremos = pd.concat([productos_mas_vendidos, productos_menos_vendidos])

# Gráficos
plt.figure(figsize=(10, 6))
sns.barplot(data=productos_extremos, x='Cantidad Vendida', y='Tienda', hue='Tipo', palette='Set2')

# Etiquetar productos en cada barra
for index, row in productos_extremos.iterrows():
    plt.text(row['Cantidad Vendida'] + 0.2,  # posición X
             index % 4 + (-0.2 if row['Tipo'] == 'Más Vendido' else 0.2),
             row['Producto'],
             color='black', fontsize=9, va='center')

plt.title('Producto Más y Menos Vendido por Tienda')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Tienda')
plt.legend(title='Tipo de Producto')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular el costo promedio de envío por tienda, redondeado a 2 decimales
envio_promedio = df_unificado.groupby('Tienda')['Costo de envío'].mean().round(2)
print("Costo de Envío Promedio por Tienda:")
display(envio_promedio.reset_index(name='Costo Promedio Envío'))
```

```python
# Gráfico de barras para envío promedio con anotaciones
plt.figure(figsize=(8, 5))

# Crear el gráfico de barras con barras más delgadas
bx = sns.barplot(x=envio_promedio.index, y=envio_promedio.values, palette='pastel', width=0.5, hue = envio_promedio.index)

# Título y etiquetas
plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Costo Promedio de Envío')
plt.xlabel('Tiendas')

# Añadir etiquetas con el valor sobre cada barra
for i, v in enumerate(envio_promedio.values):
    bx.text(i, v + 0.1, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
