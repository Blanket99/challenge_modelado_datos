# Informe Final de Análisis de Tiendas

# Introducción
El presente informe tiene como objetivo analizar el desempeño de las cuatro tiendas de Alura Store con el fin de identificar cuál de ellas presenta el peor rendimiento global y debería ser vendida. Para tomar esta decisión, se han considerado diversos indicadores clave como ingresos, volumen de ventas, calificaciones de clientes, categorías más vendidas, productos más/menos vendidos y costos de envío. Este análisis busca proporcionar una recomendación objetiva y fundamentada en datos.
# Desarrollo
## Ingresos Totales por Tienda
La Tienda 1 generó los mayores ingresos, seguida de la Tienda 2 y la Tienda 3. La Tienda 4 tuvo el ingreso total más bajo, lo que sugiere una baja rotación de productos o baja demanda. Esta es una señal de debilidad clara frente a las otras tiendas.

## Productos Más Vendidos por Tienda
En todas las tiendas, la categoría más vendida fue Muebles, lo que indica una preferencia común del mercado.

## Calificaciones Promedio
Las calificaciones de los clientes reflejan el nivel de satisfacción. Aquí, Tienda 3 obtiene el promedio más alto (4.05), mientras que Tienda 1 tiene el más bajo (3.98).

## Productos Más y Menos Vendidos
Cada tienda tiene un producto que lidera en ventas. Sin embargo, los productos menos vendidos en la Tienda 4 presentan los menores volúmenes, lo que indica que los productos con bajo desempeño no se compensan con otros de alta demanda, lo cual refleja un menor dinamismo.

## Costo Promedio de Envío
La Tienda 4 tiene el menor costo de envío promedio, lo cual es una fortaleza. Sin embargo, este punto fuerte no logra contrarrestar el bajo volumen de ventas ni los ingresos reducidos. Es decir, es eficiente logísticamente pero ineficiente comercialmente.
image.png

# Conclusión

Después de evaluar todos los factores, se concluye que la tienda que debería vender el Sr. Juan es la Tienda 4.

Justificación:
*   Es la tienda con menores ingresos totales.
*   A pesar de tener un costo logístico menor, no logra compensar el bajo volumen de ventas generales.
*   Sus calificaciones promedio no son las más altas ni las más bajas, pero tampoco destacan
*   Tiene productos menos vendidos con cantidades bajas, lo que puede reflejar baja rotación o atractivo comercial.
